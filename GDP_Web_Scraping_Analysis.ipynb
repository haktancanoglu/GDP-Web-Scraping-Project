# GDP Web Scraping Analysis

## Project Overview

This project extracts GDP data from a Wikipedia page using web scraping techniques. The data is cleaned, processed, and the top 10 largest economies are identified based on GDP.

**Technologies Used**
- Python
- Requests
- BeautifulSoup
- Pandas
- Jupyter Notebook

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
tables = soup.find_all("table")

table = tables[2]

rows = table.find_all("tr")
```

```python
data = []

for row in rows:
    cells = row.find_all("td")

    if len(cells) > 0:
        country = cells[0].text.strip()
        gdp = cells[2].text.strip()

        data.append([country, gdp])
```

```python
df = pd.DataFrame(data, columns=["Country", "GDP"])
```

```python
df["GDP"] = df["GDP"].str.replace(",", "")

df["GDP"] = pd.to_numeric(df["GDP"], errors="coerce")
```

```python
df = df.dropna()

df = df[df["Country"] != "World"]
```

```python
df = df.sort_values(by="GDP", ascending=False)
```

```python
top_10 = df.head(10).copy()

top_10["GDP"] = top_10["GDP"] / 1000
```
